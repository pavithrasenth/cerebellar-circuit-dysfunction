```python
import pandas as pd
from brian2 import *
import matplotlib.pyplot as plt

# Start fresh every time
start_scope()

# Load and clean dataset
df = pd.read_csv('ataxia_features_1.csv')
df.columns = df.columns.str.strip()

# Use any valid column, e.g., LeftStrideSpeed
signal = (df['LeftStrideSpeed'] - df['LeftStrideSpeed'].min()) / (df['LeftStrideSpeed'].max() - df['LeftStrideSpeed'].min())

# Create TimedArray input
stim = TimedArray(signal.values, dt=1*ms)

# Neuron equation
eqs2 = '''
dv/dt = (stim(t) - v) / (10*ms) : 1
'''

# Create neuron group
G2 = NeuronGroup(1, eqs2, method='euler')
G2.v = 0

# Monitor neuron
M3 = StateMonitor(G2, 'v', record=True)

# Run the simulation
run(len(df)*ms)

# Plot
plt.plot(M3.t/ms, M3.v[0])
plt.title('Neuron Simulation from LeftStrideSpeed (Brian2)')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage')
plt.grid(True)
plt.show()
```

```python
from brian2 import *
import matplotlib.pyplot as plt
import numpy as np

start_scope()
defaultclock.dt = 0.1*ms
```

```python
n_input = 10
input_group = PoissonGroup(n_input, rates=100*Hz)
```

```python
n_granule = 50
granule_eqs = '''
dv/dt = (I - v) / (10*ms) : 1
I : 1
'''
granule = NeuronGroup(n_granule, model=granule_eqs, threshold='v>1', reset='v=0', method='euler')
granule.I = 0
```

```python
input_to_granule = Synapses(input_group, granule, on_pre='v += 0.2')
input_to_granule.connect(p=0.3)
```

```python
n_purkinje = 5
purkinje_eqs = '''
dv/dt = (I - v) / (10*ms) : 1
I : 1
'''
purkinje = NeuronGroup(n_purkinje, model=purkinje_eqs, threshold='v>1', reset='v=0', method='euler')
purkinje.I = 0

granule_to_purkinje = Synapses(granule, purkinje, on_pre='v += 0.3')
granule_to_purkinje.connect(p=0.2)
```

```python
motor_eqs = '''
dv/dt = (I - v) / (10*ms) : 1
I : 1
'''
motor = NeuronGroup(1, model=motor_eqs, threshold='v>1', reset='v=0', method='euler')
motor.I = 0

purkinje_to_motor = Synapses(purkinje, motor, on_pre='v -= 0.5')
purkinje_to_motor.connect()
```

```python
mon_input = SpikeMonitor(input_group)
mon_granule = SpikeMonitor(granule)
mon_purkinje = SpikeMonitor(purkinje)
mon_motor = StateMonitor(motor, 'v', record=True)
```

```python
run(500*ms)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(4, 1, 1)
plt.plot(mon_input.t/ms, mon_input.i, 'r.')
plt.title("Input - Mossy Fibers")
plt.ylabel("Neuron Index")

plt.subplot(4, 1, 2)
plt.plot(mon_granule.t/ms, mon_granule.i, 'g.')
plt.title("Granule Cells")
plt.ylabel("Neuron Index")

plt.subplot(4, 1, 3)
plt.plot(mon_purkinje.t/ms, mon_purkinje.i, 'b.')
plt.title("Purkinje Cells (Inhibitory)")
plt.ylabel("Neuron Index")

plt.subplot(4, 1, 4)
plt.plot(mon_motor.t/ms, mon_motor.v[0])
plt.title("Simulated Motor Output")
plt.ylabel("Motor Strength")
plt.xlabel("Time (ms)")

plt.tight_layout()
plt.show()
```

```python
from brian2 import *
import matplotlib.pyplot as plt

start_scope()

granule_eqs = '''
dv/dt = (I - v) / (10*ms) : 1
I : 1
'''

G = NeuronGroup(50, granule_eqs, threshold='v>1', reset='v=0', method='euler')
G.I = 0.5

P = NeuronGroup(10, granule_eqs, threshold='v>1', reset='v=0', method='euler')
P.I = 0

S_GP = Synapses(G, P, on_pre='v_post += 0.2')
S_GP.connect(p=0.3)

# Simulate lesion
P.v[:len(P)//2] = -100

mon_P = StateMonitor(P, 'v', record=True)


net = Network(G, P, S_GP, mon_P)
net.run(500*ms)

plt.plot(mon_P.t/ms, mon_P.v[0])
plt.title("Purkinje Cell Voltage with Lesion")
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential")
plt.show()
```
